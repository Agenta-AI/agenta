## Creating a Testset

Let's create a testset with some sample data about countries and their capitals:


````python
# Create a testset with simple data
testset = await ag.testsets.acreate(
    data=[
        {"country": "Germany", "capital": "Berlin"},
        {"country": "France", "capital": "Paris"},
        {"country": "Spain", "capital": "Madrid"},
        {"country": "Italy", "capital": "Rome"},
        {"country": "Japan", "capital": "Tokyo"},
    ],
    name="Country Capitals",
    description="A testset of countries and their capitals for geography evaluation",
)

print(f"âœ… Created testset with ID: {testset.id}")
print(f"   Name: {testset.name}")
print(f"   Slug: {testset.slug}")
print(f"   Description: {testset.description}")

# Save the parent testset ID for later use
testset_id = testset.testset_id or testset.id
````

**Expected Output:**
```
âœ… Created testset with ID: 01963413-3d39-7650-80ce-3ad5d688da6c
   Name: Country Capitals
   Slug: 3ad5d688da6c
   Description: A testset of countries and their capitals for geography evaluation
```

The `acreate` function returns a `TestsetRevision` object with the following fields:
- `id`: The revision UUID
- `testset_id`: The parent testset UUID
- `version`: The revision version
- `name`: The name you provided
- `slug`: A shortened identifier
- `description`: Your description
- `data`: The test data in a structured format

## Retrieving a Testset by ID

Let's retrieve a specific testset using its ID:


````python
# Retrieve the testset we just created
retrieved_testset = await ag.testsets.aretrieve(testset_id=testset_id)

if retrieved_testset:
    print(f"\nâœ… Retrieved testset: {retrieved_testset.id}")
    print(f"   Testset ID: {retrieved_testset.testset_id}")
    print(f"   Slug: {retrieved_testset.slug}")
    print(f"   Version: {retrieved_testset.version}")

    # Access the testcases
    if retrieved_testset.data and retrieved_testset.data.testcases:
        print(f"\n   ğŸ“Š Contains {len(retrieved_testset.data.testcases)} testcase(s)")
        print("\n   Sample testcases:")
        for i, testcase in enumerate(retrieved_testset.data.testcases[:3], 1):
            print(f"      {i}. {testcase.data}")
else:
    print("âŒ Testset not found")
````

## Retrieving a Testset by Name

You can find a testset by name by filtering the results from `get_testsets`:


````python
async def get_testset_by_name(name: str):
    """Helper function to find a testset by name."""
    testsets = await ag.testsets.alist()

    if not testsets:
        return None

    for testset in testsets:
        if testset.name == name:
            return testset

    return None


# Usage example
found_testset = await get_testset_by_name("Country Capitals")

if found_testset:
    print(f"\nğŸ” Found testset by name: '{found_testset.name}'")
    print(f"   ID: {found_testset.id}")
    print(f"   Slug: {found_testset.slug}")
else:
    print("\nâŒ Testset not found")
````

```

   ID: 01963413-3d39-7650-80ce-3ad5d688da6c
   Slug: country-capitals
```
